```python
df["End_Date"]=pd.to_datetime(df["End_Date"])
```

```python
df["Start_Date"]=pd.to_datetime(df["Start_Date"])
```

```python
numeric_cols = ['Budget_USD', 'Reach', 'Impressions', 'Engagement_Rate', 'Click_Through_Rate', 'Conversions', 'ROI']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(keep='first',inplace=True)
```

```python
df.isnull()
```

```python
df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df['Budget_USD'] = df['Budget_USD'].fillna(df['Budget_USD'].median())
df['Engagement_Rate'] = df['Engagement_Rate'].fillna(df['Engagement_Rate'].mean())
df['ROI'] = df['ROI'].fillna(0)
```

```python
df.isna().sum()
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
df = df[['Campaign_ID', 'Influencer_ID', 'Campaign_Type', 'Start_Date', 'End_Date',
         'Budget_USD', 'Reach', 'Impressions', 'Engagement_Rate', 'Click_Through_Rate', 'Conversions', 'ROI']]
```

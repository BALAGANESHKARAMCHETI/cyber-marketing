```python
if 'followers' in df.columns:
    df['followers'] = df['followers'].astype(int)

```

```python
if 'followers' in df.columns:
    Q1 = df['followers'].quantile(0.25)
    Q3 = df['followers'].quantile(0.75)
    IQR = Q3 - Q1
    df = df[(df['followers'] >= Q1 - 1.5*IQR) & (df['followers'] <= Q3 + 1.5*IQR)]
```

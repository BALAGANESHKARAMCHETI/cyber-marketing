```python
import mysql.connector
import pandas as pd
from mysql.connector import Error
import numpy as np

# Database connection parameters
db_config={
  "host":"localhost",
  "user":"root",
  "password":"Bala@0505",
  "database":"cyber_marketing"
}

def get_influencers_table_to_dataframe_manual():
    """
    Connects to MySQL, retrieves product data, and stores it in a Pandas DataFrame
    by manually fetching rows and column names.
    """
    connection = None
    try:
        connection = mysql.connector.connect(**db_config)

        if connection.is_connected():
            print("Connected to MySQL successfully!")
            cursor = connection.cursor()

            query = "SELECT * FROM influencers_table"
            cursor.execute(query)

            # Fetch all rows
            rows = cursor.fetchall()

            # Get column names from cursor description
            # cursor.description returns a tuple of tuples, where each inner tuple
            # contains (column_name, type_code, display_size, internal_size, precision, scale, null_ok)
            column_names = [i[0] for i in cursor.description]

            # Create DataFrame
            df = pd.DataFrame(rows, columns=column_names)

            print("\n--- DataFrame created using cursor.fetchall() ---")
            print(df.head()) # Display first few rows
            #print(df.info())
            #print(df.describe())
            #print(df.isnull().sum())
            return df

    except Error as e:
        print(f"Error: {e}")
        return None
    finally:
        if 'cursor' in locals() and cursor is not None:
            cursor.close()
        if connection is not None and connection.is_connected():
            connection.close()
            print("MySQL connection closed.")

if __name__ == "__main__":
    df_cybermarketing_manual = get_influencers_table_to_dataframe_manual()
    if df_cybermarketing_manual is not None:
        print("\nSuccessfully retrieved data into DataFrame (Manual Method).")
```
